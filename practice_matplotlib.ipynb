```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]

fig = plt.figure()
ax = plt.subplot()
ax.plot(years, chemical_exports)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot()
ax.plot(years, chemical_exports, label = 'Chemical' , marker = 'o', ms = 8, mfc = 'white', c = 'red')
ax.plot(years, medicine_exports, label = 'Medicine',  marker = 'x', ms = 8,mfc = 'blue', c = 'blue')
plt.title('Imports of Chemicals in last decade')
ax.set_xlabel('Years')
ax.set_ylabel('Exports in Millions')

xticks = np.arange(2010, 2022, 2)
yticks = np.linspace(0.80, 1.00, 10)

plt.legend(loc='best')
plt.style.use('fivethirtyeight')
plt.grid(True)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.plot(years, chemical_exports , marker = 'o')
ax1.set_title("Chemicals Exports")
ax1.set_xlabel('Year')
ax1.set_ylabel('Sales in Million $')

ax2.plot(years, medicine_exports, marker = 'x')
ax2.set_title("Medicine Exports")
ax2.set_xlabel('Year')
ax2.set_ylabel('Sales in Million $')

plt.style.use('default')
```

```python
month = ['January', 'Feburary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [300, 320, 270, 220, 200, 140, 120, 130, 150, 190, 230, 350]

fig = plt.figure()
ax = fig.add_subplot()

ax.bar(month, sales, 0.3)
ax.set_title('Sales of Jackets')
ax.set_xlabel('Month of the Year')
ax.set_ylabel('# of units sold in a month')
plt.xticks(rotation = 45)
plt.yticks(rotation = 90)
plt.show()
```

```python
students = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
english = [85, 45, 60, 70, 83, 35]
maths = [ 80, 59, 80, 60, 65, 60]

fig, (ax1, ax2) = plt.subplots(figsize = (12,6), nrows = 1, ncols = 2)
index = np.arange(len(students))

ax1.bar(x = index-0.15, height = english, width = 0.3, label ='English', color = 'y')
ax1.bar(x = index+0.15, height = maths, width = 0.3, label = 'Math', color = 'b')
#ax1.set_xticks(ticks=index, labels = students)


ax2.barh(y = index - 0.15, width = english, height = 0.3, label = 'English', color = 'r', edgecolor = 'y')
ax2.barh(y = index + 0.15, width = maths, height = 0.3, label = 'Math', color = 'g')

plt.legend()
plt.style.use('seaborn')
```
